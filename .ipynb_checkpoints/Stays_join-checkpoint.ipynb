# Merge the airports and the hosts by city.

```python
mushed = pd.merge(hosts, airports, how='left', left_on='city', right_on='CITY' )
```

```python
print(mushed.shape)
pd.unique(mushed.IATA).size
```

Oh shit! I have got covered 18 of the 341 airports!

### Now the nosts that are not in the towns with airports have NaN in IATA column. 


## Use KKN to attach to existing airport labels

```python
toClassifyDf = mushed[mushed['IATA'].isna()==True]

toClassifyDf.reset_index(inplace=True)
```

```python
toModelDf = mushed[mushed['IATA'].isna()==False].reset_index()
```

```python
y = toModelDf['IATA']
X = toModelDf[['latitude', 'longitude']]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.20, random_state=123)
classifier = KNeighborsClassifier()
model = classifier.fit(X_train, y_train)
print('Test set model score - {:.3f}'.format(model.score(X_test, y_test)))
score = cross_val_score(model, X_train, y_train, cv=8)
print("%0.2f accuracy with a standard deviation of %0.2f" % (score.mean(), score.std()))
```

join the results back as column values.

```python
pred = model.predict(toClassifyDf[['latitude', 'longitude']])
toClassifyDf.loc[:,'IATA'] = pred
sleepovers = pd.concat([toModelDf, toClassifyDf])
# Remove min nights above 6.
print(sleepovers.minimum_nights.size)
day_hosts = sleepovers[sleepovers.minimum_nights<7]
day_hosts.minimum_nights.size
```

### Try K-means, they are distance based.
First group the cities, then add airports.

```python
from sklearn.cluster import KMeans
```

```python
X_c = toClassifyDf[['latitude','longitude']]
```

```python
K = range(2, 21)
inertia = []

for k in K:
    print("Training a K-Means model with {} clusters! ".format(k))
    print()
    kmeans = KMeans(n_clusters=k,
                    random_state=1234)
    kmeans.fit(X_c)
    inertia.append(kmeans.inertia_)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(16,8))
plt.plot(K, inertia, 'bx-')
plt.xlabel('k')
plt.ylabel('inertia')
plt.xticks(np.arange(min(K), max(K)+1, 1.0))
plt.title('Elbow Method showing the optimal k')
```

```python
clusters = KMeans(n_clusters=11, random_state=1234).fit(X_c)
```

```python
toClassifyDf['Labels'] = clusters.labels_
```
